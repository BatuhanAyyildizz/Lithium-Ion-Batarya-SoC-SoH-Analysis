```python
#Gerekli kütüphanelerin Yüklenmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Ridge
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_5=pd.read_csv('B0005.csv')
df_5=pd.DataFrame(df_5)
df_6=pd.read_csv('B0006.csv')
df_6=pd.DataFrame(df_6)
df_18=pd.read_csv('B0018.csv')
df_18=pd.DataFrame(df_18)
df7_test=pd.read_csv('B0007.csv')
df7_test=pd.DataFrame(df7_test)
```

```python
def calculate_soh(df):
    """
    Verilen veri çerçevesinde SOH (State of Health) hesaplar ve yeni bir 'SOH' sütunu ekler.
    Args:
    df (pd.DataFrame): Kapasite değerlerine sahip veri çerçevesi.

    Returns:
    pd.DataFrame: SOH sütunu eklenmiş veri çerçevesi.
    """
    df = pd.DataFrame(df)  # Ensure df is a DataFrame
    C = df['capacity'].iloc[0]  # İlk kapasite değeri
    df['SOH'] = df['capacity'] / C  # SOH kolonunu ekleme
    return df


def calculate_soc(df):
    """
    Verilen veri çerçevesinde State of Charge (SoC) hesaplar ve yeni bir 'SoC' sütunu ekler.
    Args:
    df (pd.DataFrame): Voltaj değerlerine sahip veri çerçevesi.

    Returns:
    pd.DataFrame: SoC sütunu eklenmiş veri çerçevesi.
    """
    df = pd.DataFrame(df)  # Ensure df is a DataFrame
    
    # Define the minimum and maximum voltage values
    min_voltage = df['voltage_measured'].min()  # Minimum voltage for a fully discharged battery
    max_voltage = df['voltage_measured'].max()  # Maximum voltage for a fully charged battery

    # Calculate the State of Charge (SoC) in percentage based on voltage
    df['SoC'] = ((df['voltage_measured'] - min_voltage) / (max_voltage - min_voltage)) * 100

    # Clip the SoC values to ensure they are within 0-100%
    df['SoC'] = df['SoC'].clip(0, 100)
    
    return df
```

```python
#SOH kolonlarını ekliyorum
df_5 = calculate_soh(df_5)
df_6 = calculate_soh(df_6)
df_18 = calculate_soh(df_18)
df7_test = calculate_soh(df7_test)

#SOC kolonlarını ekliyorum
df_5 = calculate_soc(df_5)
df_6 = calculate_soc(df_6)
df_18 = calculate_soc(df_18)
df7_test = calculate_soc(df7_test)
```

# LSTM modeli oluşturma ve bu modeli görselleştirme

```python
# Fonksiyonlar

# Zaman pencereleri oluşturma fonksiyonu
def create_sequences(X, y, seq_length):
    sequences = []
    targets = []
    for i in range(len(X) - seq_length):
        sequences.append(X[i:i + seq_length])
        targets.append(y[i + seq_length])
    return np.array(sequences), np.array(targets)

# Model oluşturma ve eğitme fonksiyonu
def train_model(df, seq_length=8, epochs=10, batch_size=32):
    df['datetime'] = pd.to_datetime(df['datetime'])
    
    features = ['cycle', 'ambient_temperature', 'voltage_measured', 'current_measured', 'temperature_measured', 'current_load', 'voltage_load', 'SoC']
    target = 'SOH'

    X = df[features].values
    y = df[target].values

    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()

    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

    X_seq, y_seq = create_sequences(X_scaled, y_scaled, seq_length)

    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(X_seq.shape[1], X_seq.shape[2])))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    history = model.fit(X_seq, y_seq, epochs=epochs, batch_size=batch_size, validation_split=0.1, verbose=1)

    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])
    plt.show()
    
    return model, history, scaler_X, scaler_y
```

```python
# Verisetleri
dataframes = [df_5, df_6, df_18]
models = []
histories = []
scalers_X = []
scalers_y = []

# Her bir veri seti için modeli eğitme
for i, df in enumerate(dataframes):
    print(f"Training on dataframe {i+1}")
    model, history, scaler_X, scaler_y = train_model(df)
    models.append(model)
    histories.append(history)
    scalers_X.append(scaler_X)
    scalers_y.append(scaler_y)

```

```python
# Test verisini hazırlama
df_test = df7_test  # Test verinizi buraya yükleyin
df_test['datetime'] = pd.to_datetime(df_test['datetime'])
features = ['cycle', 'ambient_temperature', 'voltage_measured', 'current_measured', 'temperature_measured', 'current_load', 'voltage_load', 'SoC']
X_test = df_test[features].values

# Tahminleri birleştirme ve ortalama alma
predictions = []

for i, model in enumerate(models):
    X_test_scaled = scalers_X[i].transform(X_test)
    seq_length = 8  # Eğitimde kullanılan pencere uzunluğu
    X_test_seq, _ = create_sequences(X_test_scaled, np.zeros(len(X_test_scaled)), seq_length)
    y_pred_scaled = model.predict(X_test_seq)
    y_pred = scalers_y[i].inverse_transform(y_pred_scaled)
    predictions.append(y_pred)
```

```python
# Ağırlıklı ortalama
weights = [0.375, 0.375, 0.25]  # Her model için ağırlıklar
weighted_predictions = np.average(predictions, axis=0, weights=weights)

# Blending modeli için veriyi hazırlama
combined_predictions = np.hstack(predictions)  # 3D'yi 2D'ye dönüştürme

# Gerçek SOH değerlerini `y_test` olarak belirtmeniz gerekiyor
y_test = df_test['SOH'].values[seq_length:]  # Gerçek SOH değerleri, test verisinin tahmin yapılabilecek kısmı

# Blending modelini eğitme
blending_model = Ridge()
blending_model.fit(combined_predictions, y_test)  # Gerçek SOH değerleri ile eğitim

# Tahminler
final_predictions = blending_model.predict(combined_predictions)
```

```python
# Gerçek SOH ve tahminlerin 0.75'e ilk geldiği noktaları bulma
threshold = 0.75

# Gerçek SOH değerlerinde threshold'a eşit veya üstü ilk noktayı bulma
real_soh_above_threshold = np.where(df_test['SOH'].values[seq_length:] <= threshold)[0]
if len(real_soh_above_threshold) > 0:
    first_real_soh_index = real_soh_above_threshold[0]
else:
    first_real_soh_index = None

# Tahminlerde threshold'a eşit veya üstü ilk noktayı bulma
predictions_above_threshold = np.where(final_predictions <= threshold)[0]
if len(predictions_above_threshold) > 0:
    first_pred_index = predictions_above_threshold[0]
else:
    first_pred_index = None

# İlgili cycle numaralarını alma
if first_real_soh_index is not None:
    real_soh_cycle = df_test['cycle'].values[seq_length:][first_real_soh_index]
else:
    real_soh_cycle = None

if first_pred_index is not None:
    pred_cycle = df_test['cycle'].values[seq_length:][first_pred_index]
else:
    pred_cycle = None
```

```python
# Model ve 
plt.figure(figsize=(12, 6))
plt.plot(df_test['cycle'].values[seq_length:], df_test['SOH'].values[seq_length:], label='Gerçek SOH', color='blue')  # Gerçek SOH değerleri
plt.plot(df_test['cycle'].values[seq_length:], final_predictions, label='Blending Model Prediction', color='red')  # Tahminler
plt.axhline(y=threshold, color='grey', linestyle='--', label='Threshold 0.75')
if real_soh_cycle is not None:
    plt.scatter(real_soh_cycle, 0.75, color='green', label='Gerçek SOH 0.75 Noktası', zorder=5)
if pred_cycle is not None:
    plt.scatter(pred_cycle, 0.75, color='orange', label='Tahmin 0.75 Noktası', zorder=5)
plt.title('Gerçek SOH ve Blending Model Tahminleri (Cycle)')
plt.xlabel('Cycle')
plt.ylabel('SOH')
plt.legend()
plt.show()

# Noktaların `cycle` numaralarını yazdırma
print(f"Gerçek RUL Cycle={real_soh_cycle}")
print(f"Tahmin RUL Cycle={pred_cycle}")
```

```python
# Gerçek ve tahmin edilen SOH değerlerini al
y_true = df_test['SOH'].values[seq_length:]  # Gerçek SOH değerleri
y_pred = final_predictions  # Tahmin edilen SOH değerleri

# MSE hesaplama
mse = mean_squared_error(y_true, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

# R² hesaplama
r2 = r2_score(y_true, y_pred)
print(f"R-squared (R²): {r2}")
```
